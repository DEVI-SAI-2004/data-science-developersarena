```python
from scipy import stats
import numpy as np
strategy_A = np.array([0.05, 0.06, 0.04, 0.07, 0.05, 0.06, 0.05])
strategy_B = np.array([0.08, 0.07, 0.09, 0.07, 0.08, 0.07, 0.09])
print("--- Comparing Business Strategies (Conversion Rates) ---")
print(f"Mean A: {np.mean(strategy_A):.4f}")
print(f"Mean B: {np.mean(strategy_B):.4f}")
t_stat, p_value = stats.ttest_ind(strategy_B, strategy_A)
print(f"\nT-statistic: {t_stat:.4f}")
print(f"P-value: {p_value:.4f}")
alpha = 0.05
if p_value < alpha:
  print(f"\nConclusion: Reject the Null Hypothesis (p < {alpha}). There is a statistically significant difference.")
else:
  print(f"\nConclusion: Fail to Reject the Null Hypothesis (p >= {alpha}). There is no statistically significant difference.")
```

## Week 6: Statistical Analysis Report

### Objective
Comparing the effectiveness of two business strategies (A and B) using an independent two-sample t-test.

### Statistical Test Details
Test Used:T-Test.
* **Null Hypothesis (H₀):** There is no significant difference in performance (mean) between Strategy A and Strategy B.
* **Significance Level (α):** 0.05.

### Results
* **Strategy A Mean:** 0.0543
* **Strategy B Mean:** 0.0786
* **T-statistic:** 5.0990
* **P-value:** 0.0006

### Conclusion
Since the P-value (0.0006) is much lower than the significance level ($\alpha = 0.05$), we **reject the Null Hypothesis**. This confirms that the difference in means is statistically significant. **Strategy B is proven to be statistically more effective** than Strategy A.

```python
#Client Project:House Price Prediction Model
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, r2_score
import pandas as pd
import numpy as np
import joblib
data = {
    'Size_SqFt': [1000, 1500, 2000, 2500, 3000, 1200, 1800, 2800],
    'Price_k': [200, 250, 350, 420, 500, 210, 310, 480]
}
df = pd.DataFrame(data)
X = df[['Size_SqFt']]
y = df['Price_k']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
model = LinearRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print("--- Linear Regression Model Results ---")
print(f"Trained Model Coefficients: {model.coef_}")
print(f"Mean Absolute Error (MAE): {mae:.2f} k")
print(f"R-squared (R2) Score: {r2:.4f}")
joblib.dump(model, 'house_price_model.joblib')
print("\nModel saved successfully as 'house_price_model.joblib'.")
```

```python
from sklearn.model_selection import GridSearchCV, cross_val_score
from sklearn.tree import DecisionTreeRegressor
from sklearn.datasets import make_regression
import numpy as np
X_dummy, y_dummy = make_regression(n_samples=50, n_features=5, random_state=42)
dt_model = DecisionTreeRegressor(random_state=42)
param_grid = {
    'max_depth': [3, 5, 7, 9],
    'min_samples_leaf': [1, 5, 10]
}
grid_search = GridSearchCV(estimator=dt_model, param_grid=param_grid, cv=3, scoring='neg_mean_squared_error')
grid_search.fit(X_dummy, y_dummy)
print("--- Model Tuning Results (GridSearchCV) ---")
print(f"Best Hyperparameters Found: {grid_search.best_params_}")
print(f"Best Score (Negative MSE): {grid_search.best_score_:.2f}")
best_dt = grid_search.best_estimator_
cv_scores = cross_val_score(best_dt, X_dummy, y_dummy, cv=5, scoring='r2')

print(f"\nFinal 5-Fold Cross-Validation R-squared Scores: {cv_scores}")
print(f"Average R-squared (Evaluation Metric): {np.mean(cv_scores):.4f}")
```
