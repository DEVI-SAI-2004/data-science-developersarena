```python
temperatures = [25.5, 27.0, 24.5, 26.5, 28.0, 23.5, 26.0]
total_temp = 0
count = 0
for temp in temperatures:
    total_temp += temp
    count += 1
if count > 0:
    average_temp = total_temp / count
    print(f"The recorded temperatures are: {temperatures}")
    print(f"The total number of readings is: {count}")
    print(f"The average temperature for the period is: {average_temp:.2f}°C")
else:
    print("No temperature data available to calculate the average.")
```

```python
customer_ids_raw = [101, 105, 103, 102, 101, 104, 105, 106, 100]
def clean_and_filter_ids(data):
  unique_ids = set(data)
  unique_list = list(unique_ids)
  filtered_ids = [id for id in unique_list if id > 102]
  return unique_list, filtered_ids
unique_ids, high_value_ids = clean_and_filter_ids(customer_ids_raw)
print(f"Raw Customer IDs: {customer_ids_raw}")
print(f"Unique Customer IDs: {unique_ids}")
print(f"Filtered IDs (>102): {high_value_ids}")
```

```python
import numpy as np
import pandas as pd
data = {
    'Product': ['A', 'B', 'A', 'C', 'B', 'C', 'A', 'B'],
    'Sales': [150, 200, 155, np.nan, 210, 300, 160, 205],
    'Region': ['East', 'West', 'East', 'North', 'West', 'East', 'North', 'West']
}
df = pd.DataFrame(data)
print("Original DataFrame:")
print(df)
print("-" * 30)
df_cleaned = df.dropna()
print("Cleaned DataFrame:")
print(df_cleaned)
print("-" * 30)
average_sales = df_cleaned.groupby('Product')['Sales'].mean().reset_index()
print("Aggregated(Average Sales by Product):")
print(average_sales)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
data = {
    'Engine_Size': [1.8, 2.0, 1.5, 2.5, 1.8, 3.0],
    'Horsepower': [140, 165, 110, 200, 150, 250],
    'Efficiency': [30, 28, 35, 22, 29, 20]
}
df = pd.DataFrame(data)
fig, axes = plt.subplots(1, 2, figsize=(12, 5))
plt.suptitle('Automotive Feature Dashboard Analysis', fontsize=16)
sns.scatterplot(x='Engine_Size', y='Horsepower', data=df, ax=axes[0], hue='Efficiency', size='Efficiency', sizes=(50, 200))
axes[0].set_title('Scatter Plot: Engine Size vs. Horsepower')
axes[0].grid(True)
axes[1].hist(df['Efficiency'], bins=4, edgecolor='black', alpha=0.7)
axes[1].set_title('Histogram: Distribution of Efficiency')
axes[1].set_xlabel('Efficiency (MPG)')
axes[1].set_ylabel('Frequency')
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.savefig('Dashboard_plots.png')
plt.show()
```
